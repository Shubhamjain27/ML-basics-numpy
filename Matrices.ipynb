Matrix Multiplication:

1. Haddamard Product(Element-wise Multiplication) - Used in Convolutions
2. Matmul - Matrix multiplication used for performing linear transformations one after the other.

```python
#matrix operations
# (n,d)..n: no. of elements, d:dimensions

np.random.seed(100)
X1 = np.random.rand(4,7)
X2 = np.random.rand(4,7)

#Matmul
X3 = np.matmul(X1,X2.T)
print(X3.shape)


#Hadammard Product
X4 = X1*X2
```
